```python
def extract_paragraphs(document_path):
    doc = docx.Document(document_path)
    paragraphs = []
    for p in doc.paragraphs:
        if p.text.strip():
            modified_paragraph = p.text.replace(".", "|")
            paragraphs.append(modified_paragraph)
    return paragraphs
```

```python
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast
model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-one-to-many-mmt")
tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-one-to-many-mmt",src_lang="en_XX")
```

```python
# Add content to the document
def write_paragraphs_to_document(paragraphs, output_path):
    doc = docx.Document()
    for paragraph in paragraphs:
        doc.add_paragraph(paragraph)

    doc.save(output_path)

output_path = '/content/output_file.docx'      #'/content/outputdraftdoc.docx'

write_paragraphs_to_document(translations, output_path)
```
